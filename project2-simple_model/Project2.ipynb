# Simple models
### target variable to y 

```python
y = housing['price']
y.head()
```

```python
y_pred = y.mean()
y_pred
```

```python
pd.np.sqrt(y_pred)
```

```python
# Calculate error
error = y - y_pred
error.head()
```

```python
# Square errors
error_squared = error ** 2
error_squared.head()
```

```python
# Sum errors
sum_of_errors = error_squared.sum()
sum_of_errors
```

```python
# Compute the mean of the errors
model_1_mse  = sum_of_errors / len(y)
model_1_mse
```

```python
# Model 2 - Median
y_pred = y.median()
print(y_pred)
model_2_mse = ((y - y_pred) ** 2).sum() / len(y)
model_2_mse
```

```python
# Model 3 - Max
y_pred = y.max()
print(y_pred)
model_3_mse = ((y - y_pred) ** 2).sum() / len(y)
model_3_mse
```

```python
# Model 4 - M
y_pred = y.min()
print(y_pred)
model_4_mse = ((y - y_pred) ** 2).sum() / len(y)
model_4_mse
```

```python
# Model 5 - Standard deviation
y_pred = y.std()
print(y_pred)
model_5_mse = ((y - y_pred) ** 2).sum() / len(y)
model_5_mse
```

```python
# Plot errors for simple models
errors = {
    'model_1_mse': model_1_mse,
    'model_2_mse': model_2_mse, 
    'model_4_mse': model_4_mse, 
    'model_5_mse': model_5_mse
}

pd.Series(errors).plot(kind='bar', figsize=(20, 15), rot=0);
```
